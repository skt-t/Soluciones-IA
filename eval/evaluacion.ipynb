```python
import os
import wikipedia
from langchain_openai import ChatOpenAI

treintaysiete = """

â–‘â–’â–“â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–“â–’â–‘â–‘â–’â–“â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–“â–’â–‘ 
       â–‘â–’â–“â–ˆâ–“â–’â–‘â–’â–“â–ˆâ–“â–’â–‘â–‘â–’â–“â–ˆâ–“â–’â–‘ 
       â–‘â–’â–“â–ˆâ–“â–’â–‘      â–‘â–’â–“â–ˆâ–“â–’â–‘ 
â–‘â–’â–“â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–“â–’â–‘      â–‘â–’â–“â–ˆâ–“â–’â–‘  
       â–‘â–’â–“â–ˆâ–“â–’â–‘     â–‘â–’â–“â–ˆâ–“â–’â–‘  
       â–‘â–’â–“â–ˆâ–“â–’â–‘    â–‘â–’â–“â–ˆâ–“â–’â–‘   
â–‘â–’â–“â–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–ˆâ–“â–’â–‘     â–‘â–’â–“â–ˆâ–“â–’â–‘   


"""

# Configurar el idioma de Wikipedia
wikipedia.set_lang("es")

# ConfiguraciÃ³n del LLM
try:
    llm = ChatOpenAI(
        model="gpt-4o",
        openai_api_base=os.environ.get("GITHUB_BASE_URL"),
        openai_api_key=os.environ.get("GITHUB_TOKEN"),
        temperature=0
    )
    print("âœ… LLM de LangChain configurado.")
    print(treintaysiete)
except Exception as e:
    print(f"âŒ Error configurando el LLM: {e}")
    llm = None
```

```python
from langchain.agents import tool, create_openai_tools_agent, AgentExecutor
from langchain import hub

@tool
def get_wikipedia_summary(query: str) -> str:
    """Busca en Wikipedia un tema y devuelve un resumen de 2 frases. Ãštil para obtener informaciÃ³n sobre personas, lugares o conceptos."""
    try:
        return wikipedia.summary(query, sentences=2)
    except Exception as e:
        return f"OcurriÃ³ un error: {e}"

tools = [get_wikipedia_summary]

prompt = hub.pull("hwchase17/openai-tools-agent")

agent = create_openai_tools_agent(llm, tools, prompt)
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)

print("âœ… Agente y herramientas listos.")
```

```python
from langchain.memory import ConversationSummaryMemory

# Esta memoria necesita un LLM para hacer los resÃºmenes
memory_summary = ConversationSummaryMemory(llm=llm, memory_key="chat_history", return_messages=True)

def chat_with_agent_summary(query: str):
    # busca el historial de la memoria
    history = memory_summary.load_memory_variables({})["chat_history"]
    # respuesta en base a la memoria
    response = agent_executor.invoke({"input": query, "chat_history": history})
    # guarda la nueva interaccion en la memoria
    memory_summary.save_context({"input": query}, {"output": response["output"]})
    return response["output"]
```

```python
print("Pregunta 1: Necesito saber como encontrar una buena tienda para hacerme un piercing")
chat_with_agent_summary("Necesito saber como encontrar una buena tienda para hacerme un piercing")

print("Pregunta 2: Â¿que define a una buena tienda de piercings?")
chat_with_agent_summary("Â¿que define a una buena tienda de piercings?")

print("Pregunta 3: Â¿que piercing me recomiendas para una primera vez?")
chat_with_agent_summary("Â¿que piercing me recomiendas para una primera vez?")

print("Pregunta 4: Â¿como debo cuidar este piercing?")
chat_with_agent_summary("Â¿como debo cuidar este piercing?")

print("Pregunta 5: Â¿en que tipo de trabajos me aceptarian un piercing?")
chat_with_agent_summary("Â¿en que tipo de trabajos me aceptarian un piercing?")

print("Pregunta 3: recuerdas que piercing me recomendaste?")
response1_summary = chat_with_agent_summary("Â¿Sobre quÃ© estÃ¡bamos hablando?")
print(f"Respuesta 1: {response1_summary}")
```
